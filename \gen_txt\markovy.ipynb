```
INFILE = "Valeriy.txt"
OUTFILE = "output.txt"
with open(INFILE, encoding="UTF-") as infile, open(OUTFILE, "w", encoding="UTF-8") as outfile:
    for line in infile:
        if line!='\n':
            outfile.write(line)
```
